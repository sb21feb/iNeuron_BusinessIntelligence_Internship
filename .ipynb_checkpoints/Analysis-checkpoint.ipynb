#### Dropping Duplicates

```python
google_play_store.drop_duplicates(subset = 'App', inplace = True)
```

```python
# Modify the 'Reviews' data and change the type 'object' to 'float'
reviews = [i for i in google_play_store['Reviews']]

def modify_reviews(reviews_list):
    """
    This function is replacing 'M' in reviews data with million
    """
    cleaned_data = []
    for review in reviews_list:
        if 'M' in review:
            review = review.replace('M', '')
            review = float(review) * 1000000  # 1M = 1,000,000
        cleaned_data.append(review)
    return cleaned_data

google_play_store['Reviews'] = modify_reviews(reviews)
google_play_store['Reviews'] = google_play_store['Reviews'].astype(float)
```

```python
# Modify the 'Size' data and change the type 'object' to 'float'

# found value with '1,000+' in one of record, remove it from data_frame as uncertain whether it is 'M' or 'k'
index = google_play_store[google_play_store['Size'] == '1,000+'].index
google_play_store.drop(axis=0, inplace=True, index=index)

sizes = [i for i in google_play_store['Size']]

def modify_sizes(sizes_list):
    """
    As sizes are represented in 'M' and 'k', we remove 'M'
    and convert 'k'/kilobytes into megabytes
    """
    modified_data = []
    for size in sizes_list:
        if 'M' in size:
            size = size.replace('M', '')
            size = float(size)
        elif 'k' in size:
            size = size.replace('k', '')
            size = float(size)
            size = size/1024  # 1 megabyte = 1024 kilobytes
        # representing 'Varies with device' with value 0
        elif 'Varies with device' in size:
            size = float(0)
        modified_data.append(size)
    return modified_data

google_play_store['Size'] = modify_sizes(sizes)
google_play_store['Size'] = google_play_store['Size'].astype(float)
```

```python
#clean the 'Installs' data and change the type 'object' to 'float'
installs = [i for i in google_play_store['Installs']]

def modify_installs(installs_list):
    modified_data = []
    for install in installs_list:
        if ',' in install:
            install = install.replace(',', '')
        if '+' in install:
            install = install.replace('+', '')
        install = int(install)
        modified_data.append(install)
    return modified_data
        
google_play_store['Installs'] = modify_installs(installs)
google_play_store['Installs'] = google_play_store['Installs'].astype(float)
```

```python
# Modify the 'Price' data and change the type 'object' to 'float'
prices = [i for i in google_play_store['Price']]

def modify_prices(prices_list):
    modified_data = []
    for price in prices_list:
        if '$' in price:
            price = price.replace('$', '')
        modified_data.append(price)
    return modified_data

google_play_store['Price'] = modify_prices(prices)
google_play_store['Price'] = google_play_store['Price'].astype(float)
```

```python
def table_info(conn, cursor):
    """
    prints out all of the columns of every table in the DB
    
    conn:database connection object
    cursor:cursor object
    """
    tables = cursor.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
    for table_name in tables:
        table_name = table_name[0]
        table = pd.read_sql_query("Select * from {} LIMIT 0".format(table_name),conn)
        print(table_name)
        for col in table.columns:
            print("\t"+col)
        print()
```

```python
#Printing all the columns present in database
table_info(conn,cur)
```

```python
# Function to group the apps based on its size in MB

def size_apps(var):
  '''
  This function groups the size of an app 
  between ~0 to 100 MB into certain intervals.
  '''
  try:
    if var < 1:
      return 'Below 1'
    elif var >= 1 and var <10:
      return '1-10'
    elif var >= 10 and var <20:
      return '10-20'
    elif var >= 20 and var <30:
      return '20-30'
    elif var >= 30 and var <40:
      return '30-40'
    elif var >= 40 and var <50:
      return '40-50'
    elif var >= 50 and var <60:
      return '50-60'
    elif var >= 60 and var <70:
      return '60-70'
    elif var >= 70 and var <80:
      return '70-80'
    elif var >= 80 and var <90:
      return '80-90'
    else:
      return '90 and above'
  except:
    return var
```

```python
google_play_store['size_group']=google_play_store['Size'].apply(lambda x : size_apps(x))
google_play_store.head()
```

### **`Size effect on Installs`**

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='size_group',y='Installs', data=google_play_store)
plt.title('Size Versus Installs')
plt.xticks(rotation=90)
plt.xlabel('Size')
plt.ylabel('Number of Installs')
plt.show()
```

#### Findings:
Maximum number of installs are made for Below 1 size group. Minimum number of installs are made for 1-10 size group.

```python
data = google_play_store[google_play_store['size_group']=='Below 1']['Type'].value_counts()
labels = ['Free','Paid']
plt.figure(figsize=(10,10))
explode=(0.1,0.1)
plt.pie(data,labels=labels, explode=explode,autopct='%.2f%%',textprops={'fontsize': 15})
plt.title('Below 1',loc='center')
plt.legend()
```

#### Findings:
As we know that maximum number of installs were being made in Below 1 size group, we can infer from above pie chart that 90.96% are Free and 9.04% are paid. So, both size and type are playing crucial role here in the number of installs.

```python
data = google_play_store[google_play_store['size_group']=='1-10']['Type'].value_counts()
labels = ['Free','Paid']
plt.figure(figsize=(10,10))
explode=(0.1,0.1)
plt.pie(data,labels=labels, explode=explode,autopct='%.2f%%',textprops={'fontsize': 15})
plt.title('1-10',loc='center')
plt.legend()
```

#### Findings:
As we know that minimum number of installs were being made in 1-10 size group, we can infer from above pie chart that 91.88% are Free and 8.12% are paid. Although the Paid percentage is lower than Free one, we can stil say that Paid type is playing some role here for minimum installs for this size group as there are other size groups larger than 1-10 but having more installs.
